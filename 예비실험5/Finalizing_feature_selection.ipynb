```python
import pandas as pd
```

```python
alsfrs_raw = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/예비실험 5/ALSFRS_original.csv')
X_and_y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/예비실험 5/feature_and_target_for_MI.csv')
alsfrs_tot_slope = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/예비실험 5/ALSFRS_slope.csv')
```

### 1. Adding ALSFRS_Q features

```python
extent = list(alsfrs_tot_slope["SubjectID"].unique())
```

```python
alsfrs_raw.query("SubjectID == {0}".format(extent), inplace=True)
```

```python
alsfrs_raw.query("feature_delta < 92", inplace=True)
```

```python
alsfrs_raw.drop(columns=['Q5a_Cutting_without_Gastrostomy', 'Q5b_Cutting_with_Gastrostomy'], inplace=True)
```

```python
alsfrs_raw['Bulbar'] = alsfrs_raw['mouth'] - alsfrs_raw['Q2_Salivation']
```

```python
alsfrs_raw['motor'] = alsfrs_raw['hands'] + alsfrs_raw['trunk'] + alsfrs_raw['leg']
```

1-1. mean value

```python
df = alsfrs_raw.groupby('SubjectID').agg('mean')
df.drop(columns=['feature_delta','mouth', 'hands', 'leg', 'trunk', 'respiratory'], inplace=True)
```

```python
df = df.reset_index()
```

```python
df.columns=['SubjectID', 'mean_alsfrs_total', 'mean_Q1_Speech', 'mean_Q2_Salivation', 'mean_Q3_Swallowing', 'mean_Q4_Handwriting', 'mean_Q5_Cutting', 'mean_Q6_Dressing_and_Hygiene', 'mean_Q7_Turning_in_Bed', 'mean_Q8_Walking', 'mean_Q9_Climbing_Stairs', 'mean_Q10_Respiratory', 'mean_Bulbar', 'mean_motor']
```

```python
mean_alsfrs_scores = df.copy()
mean_alsfrs_scores = mean_alsfrs_scores.sort_values(by=['SubjectID'], axis=0, ascending=True)
mean_alsfrs_scores
```

1-2. min value

```python
df2 = alsfrs_raw.groupby('SubjectID').agg('min')
df2.drop(columns=['feature_delta','mouth', 'hands', 'leg', 'trunk', 'respiratory'], inplace=True)
```

```python
df2 = df2.reset_index()
```

```python
df2.columns=['SubjectID', 'min_alsfrs_total', 'min_Q1_Speech', 'min_Q2_Salivation', 'min_Q3_Swallowing', 'min_Q4_Handwriting', 'min_Q5_Cutting', 'min_Q6_Dressing_and_Hygiene', 'min_Q7_Turning_in_Bed', 'min_Q8_Walking', 'min_Q9_Climbing_Stairs', 'min_Q10_Respiratory', 'min_Bulbar', 'min_motor']
```

```python
min_alsfrs_scores = df2.copy()
min_alsfrs_scores = min_alsfrs_scores.sort_values(by=['SubjectID'], axis=0, ascending=True)
min_alsfrs_scores
```

1-3. slope

```python
df3 = alsfrs_raw.groupby('SubjectID').agg(['first', 'last'])
df3['interval'] = df3.iloc[:, 1] - df3.iloc[:, 0]
df3['slope_Q1_Speech'] = (df3.iloc[:,5] - df3.iloc[:,4])/df3['interval']
df3['slope_Q2_Salivation'] = (df3.iloc[:,7] - df3.iloc[:,6])/df3['interval']
df3['slope_Q3_Swallowing'] = (df3.iloc[:,9] - df3.iloc[:,8])/df3['interval']
df3['slope_Q4_Handwriting'] = (df3.iloc[:,11] - df3.iloc[:,10])/df3['interval']
df3['slope_Q5_Cutting'] = (df3.iloc[:,13] - df3.iloc[:,12])/df3['interval']
df3['slope_Q6_Dressing_and_Hygiene'] = (df3.iloc[:,15] - df3.iloc[:,14])/df3['interval']
df3['slope_Q7_Turning_in_Bed'] = (df3.iloc[:,17] - df3.iloc[:,15])/df3['interval']
df3['slope_Q8_Walking'] = (df3.iloc[:,19] - df3.iloc[:,18])/df3['interval']
df3['slope_Q9_Climbing_Stairs'] = (df3.iloc[:,21] - df3.iloc[:,20])/df3['interval']
df3['slope_Q10_Respiratory'] = (df3.iloc[:,23] - df3.iloc[:,22])/df3['interval']
df3['slope_Bulbar'] = (df3.iloc[:,35] - df3.iloc[:,34])/df3['interval']
df3['slope_motor'] = (df3.iloc[:,37] - df3.iloc[:,36])/df3['interval']
```

```python
df4 = df3.iloc[:, 39:]
df4 = df4.reset_index()
```

```python
slope_alsfrs_scores = df4.copy()
slope_alsfrs_scores = slope_alsfrs_scores.sort_values(by=['SubjectID'], axis=0, ascending=True)
slope_alsfrs_scores
```

```python
alsfrs_tot_slope = alsfrs_tot_slope.sort_values(by=['SubjectID'], axis=0, ascending=True)
alsfrs_tot_slope = alsfrs_tot_slope.reset_index()
```

```python
df1 = mean_alsfrs_scores.merge(min_alsfrs_scores, on='SubjectID', how='outer')
df2 = slope_alsfrs_scores.iloc[:,1:]
df3 = alsfrs_tot_slope.iloc[:, 2]

alsfrs_q_mean_min_slope = pd.concat([df1, df2, df3], axis=1)
alsfrs_q_mean_min_slope.columns = ['SubjectID','mean_alsfrs_total','mean_Q1_Speech','mean_Q2_Salivation','mean_Q3_Swallowing','mean_Q4_Handwriting','mean_Q5_Cutting','mean_Q6_Dressing_and_Hygiene',
'mean_Q7_Turning_in_Bed','mean_Q8_Walking','mean_Q9_Climbing_Stairs','mean_Q10_Respiratory','mean_Bulbar','mean_motor','min_alsfrs_total','min_Q1_Speech','min_Q2_Salivation','min_Q3_Swallowing',
'min_Q4_Handwriting','min_Q5_Cutting','min_Q6_Dressing_and_Hygiene','min_Q7_Turning_in_Bed','min_Q8_Walking','min_Q9_Climbing_Stairs','min_Q10_Respiratory','min_Bulbar','min_motor','slope_Q1_Speech',
'slope_Q2_Salivation','slope_Q3_Swallowing','slope_Q4_Handwriting','slope_Q5_Cutting','slope_Q6_Dressing_and_Hygiene','slope_Q7_Turning_in_Bed','slope_Q8_Walking','slope_Q9_Climbing_Stairs','slope_Q10_Respiratory',
'slope_Bulbar','slope_motor','ALSFRS_slope']
alsfrs_q_mean_min_slope
```

### 2. Selecting features

```python
X_and_y.drop(columns='Unnamed: 0', inplace=True)
```

```python
# Add a meta feature
X_and_y['diag_minus_onset'] = X_and_y['diag_delta'] - X_and_y['onset_delta']
```

```python
X_and_y_mean_version = X_and_y[['Age', 'Gender', 'onset_site', 'onset_delta', 'diag_delta', 'diag_minus_onset', 'mean_alsfrs_total' ,'alsfrs_total_slope', 'fvc_slope', 'fvc_mean']]
X_and_y_min_version = X_and_y[['Age', 'Gender', 'onset_site', 'onset_delta', 'diag_delta', 'diag_minus_onset', 'min_alsfrs_total' ,'alsfrs_total_slope', 'fvc_slope', 'fvc_min']]
```

```python
df_mean = mean_alsfrs_scores.iloc[:, 2:]
df_min = min_alsfrs_scores.iloc[:, 2:]
id_list = alsfrs_q_mean_min_slope[['SubjectID']]
```

```python
mean_ver_features_and_target = pd.concat([id_list, X_and_y_mean_version, df_mean, df3], axis=1)
min_ver_features_and_target = pd.concat([id_list, X_and_y_min_version, df_min, df3], axis=1)
```

```python
Gend = {"F":0, "M":1}
os = {"Other":0, "Bulbar":1, "Limb":2, "Limb and Bulbar":3}
```

```python
mean_ver_features_and_target = mean_ver_features_and_target.replace({'Gender':Gend})
min_ver_features_and_target = min_ver_features_and_target.replace({'Gender':Gend})
```

```python
mean_ver_features_and_target = mean_ver_features_and_target.replace({'onset_site':os})
min_ver_features_and_target = min_ver_features_and_target.replace({'onset_site':os})
```
